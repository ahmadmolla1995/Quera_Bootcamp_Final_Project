```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd

file_path = '/content/drive/MyDrive/AI/deep/second_project/secondpart/train_data_sample_spell_checked_clean.csv'

df = pd.read_csv(file_path)
df.head()
```

```python
# فیلتر کردن نظراتی که درباره گارانتی هستند
guarantee_reviews = df[df['classification'] == 'guarantee']

# محاسبه میانگین امتیاز برای هر محصول
average_rating_per_product = guarantee_reviews.groupby('asin')['overall'].mean().reset_index()

average_rating_per_product.columns = ['asin', 'average_overall_guarantee']
average_rating_per_product
```

```python
# فیلتر کردن کالاهایی که در هر دو حالت گارانتی‌دار و بدون گارانتی موجود هستند

guarantee_counts = df.groupby(['asin', 'classification']).size().unstack().fillna(0)
dual_status_products = guarantee_counts[(guarantee_counts['guarantee'] > 0) & (guarantee_counts['no guarantee'] > 0)].index
df_dual_status = df[df['asin'].isin(dual_status_products)]

# میانگین امتیاز
comparison = df_dual_status.groupby(['asin', 'classification'])['overall'].mean().unstack()
print(comparison)


# میانگین رضایت برای حالت گارانتی‌دار و بدون گارانتی
average_satisfaction = comparison.mean()
avg_with_guarantee = average_satisfaction['guarantee']
avg_without_guarantee = average_satisfaction['no guarantee']

print(f"میانگین رضایت با گارانتی: {avg_with_guarantee:.2f}")
print(f"میانگین رضایت بدون گارانتی: {avg_without_guarantee:.2f}")

```
